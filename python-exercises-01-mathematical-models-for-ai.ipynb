## Task 1: Lists, Dictionaries, and Tuples

```python
# Task 1: Lists
nums = [3, 4, 7, 8, 15]
cubes = [x**3 for x in nums]
print("Cubes of the given list:", cubes)
```

```python
# Task 1: Dictionaries
my_dic = {}
my_dic['person'] = 2
my_dic['cat'] = 4
my_dic['spider'] = 8
my_dic['horse'] = 4

# Loop through the dictionary and print each animal with legs
for animal, legs in my_dic.items():
    print(f"{animal}: {legs} legs")

# Sum the legs
total_legs = sum(my_dic.values())
print("Total number of legs:", total_legs)
```

```python
# Task 1: Tuples
D = (1, 15, 4, [5, 10])

# Change the value in the list from 5 to 3
D[3][0] = 3
print("Modified tuple D:", D)

# Delete the tuple D
del D

# Tuple E operations
E = ('a', 'p', 'p', 'l', 'e')
print("Number of occurrences of 'p':", E.count('p'))
print("Index of 'l':", E.index('l'))
```

## Task 2: Functions and For Loops

```python
# Task 2: Compute velocity
def Compute(distance, time):
    if time != 0:
        return distance / time
    else:
        return "Time cannot be zero."

velocity = Compute(100, 2)
print("Velocity:", velocity)
```

```python
# Task 2: Sum of even numbers till 20
even = [x for x in range(2, 21, 2)]

def Sum(lst):
    total = 0
    for num in lst:
        total += num
    return total

print("Sum of even numbers:", Sum(even))
```

```python
# Task 2: Intersection of two lists
def Intersection(lst1, lst2):
    return [value for value in lst1 if value in lst2]

list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]
print("Intersection of lists:", Intersection(list1, list2))
```

```python
# Task 2: Merging two dictionaries
def merge_dicts(dict1, dict2):
    result = dict1.copy()
    for key, value in dict2.items():
        if key in result:
            result[key] += value
        else:
            result[key] = value
    return result

a = {'one': 12, 'two': 10, 'three': 22}
b = {'four': 14, 'one': 10, 'three': 2}
merged = merge_dicts(a, b)
print("Merged dictionary:", merged)
```

```python
# Task 2: Filter odd values
def filter_odd_values(input_dict):
    return {k: v for k, v in input_dict.items() if v % 2 != 0}

my_dict = {'a': 13, 'b': 12, 'c': 5, 'd': 3, 'e': 29}
odd_dict = filter_odd_values(my_dict)
print("Filtered dictionary with odd values:", odd_dict)
```

```python
# Task 2: Generate lists and add them element-wise
odd_numbers = [x for x in range(1, 25) if x % 2 != 0][:12]
multiples_of_3 = [x for x in range(3, 37, 3)]

sum_list = [odd + multiple for odd, multiple in zip(odd_numbers, multiples_of_3)]
print("Element-wise addition of two lists:", sum_list)
```

## Task 3: Numpy Operations

```python
import numpy as np

# Task 3: Numpy operations
M = np.array([[10, 11, 12], [14, 15, 16], [18, 19, 20]])
z = np.array([1, 0, 1])

# Convert M to numpy array (already done above)
# Slicing to get subarray
b = M[:2, :2]
print("Subarray b:", b)

# Create empty matrix y with the same shape as M
y = np.empty_like(M)

# Add z to each column of M
for i in range(M.shape[1]):
    y[:, i] = M[:, i] + z

print("Matrix after adding z to each column:", y)
```

## Task 4: Pandas DataFrame

```python
import pandas as pd

# Task 4: Creating DataFrame
data = {'F1': [1, 2, 3, 4, 5],
        'F2': [6, 7, 9, 8, 2],
        'F3': [81, 90, 94, 92, 88],
        'F4': [7, 700, 900, 60, 4]}

df = pd.DataFrame(data)

# Print the first two rows
print("First two rows:\n", df.head(2))

# Print the second column
print("Second column:\n", df['F2'])

# Rename third column to 'Feature3'
df.rename(columns={'F3': 'Feature3'}, inplace=True)

# Add a new column 'Sum' and calculate row-wise sum
df['Sum'] = df.sum(axis=1)
print("DataFrame with 'Sum' column:\n", df)
```
